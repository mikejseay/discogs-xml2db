```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:/Users/mikejseay/Documents/Code/discogs-xml2db/output_csv/releasesPerStyle.csv')
```

```python
df.set_index('style', inplace=True)
```

```python
WC = WordCloud().generate_from_frequencies(df.to_dict()['instances'])
```
